# Exercise: Anscombe's quartet

```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('data/anscombe_datasets.csv')
df.head()
```

```python
alt.Chart(df).mark_circle().encode(
    x='X:Q',
    y='Y:Q',
    facet=alt.Facet('dataset:N', columns=2),
)
```

```python
# Base scatter plot
points = alt.Chart().mark_circle(
    size=60,
    opacity=0.7,
    color='#11296b'  # Colore dei punti
    ).encode(
    x=alt.X('X:Q'),
    y=alt.Y('Y:Q'),
    tooltip=['X', 'Y']
).properties(
    width=300,
    height=200
)

# Regression line - CORREZIONE QUI
regression = alt.Chart().mark_line(
    color='red', size=1, strokeDash=[5, 5]  # Linea tratteggiata
    ).transform_regression(
        "X", "Y", method="poly", order=1, as_=["X", "reg"]
    ).encode(
    x='X:Q',
    y='reg:Q'  # Utilizza Y_pred generato dalla trasformazione regression
)

# Combine the scatter plot and regression line
chart = alt.layer(points, regression, data=df).facet(
    facet=alt.Facet('dataset:N'),
    title='',
    columns=2
)

# Display the chart
chart
```
